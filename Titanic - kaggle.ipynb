```python
titles = set()
for name in df['Name']:
  titles.add(name.split(',')[1].split('.')[0].strip())
titles
```

```python
title_dictionary = {"Capt": "Officer","Col": "Officer","Major": "Officer","Jonkheer": "Royalty","Don": "Royalty","Sir" : "Royalty","Dr": "Officer","Rev": "Officer","the Countess":"Royalty","Mme": "Mrs","Mlle": "Miss","Ms": "Mrs","Mr" : "Mr","Mrs" : "Mrs","Miss" : "Miss","Master" : "Master","Lady" : "Royalty"}
```

```python
df['title'] = df['Name'].map(lambda name: name.split(',')[1].split('.')[0].strip())

df['title'] = df['title'].map(title_dictionary)

df
```

```python
df_1 = df.drop(['Name', 'Ticket','Cabin', 'PassengerId', 'Died'], axis= 1 )
df_1
```

```python
df_1.Sex.value_counts()
```

```python
df_1.Sex = df_1.Sex.map({'female': 0, 'male': 1 })

df_1.Embarked = df_1.Embarked.map({'S':0, 'C':1, 'Q':2, 'nan': 'NaN'})
df_1.title=df_1.title.map({'Mr':0, 'Miss':1, 'Mrs':2,'Master':3,'Officer':4,'Royalty':5})

df_1.head()
```

```python
df_1.isnull().sum()
```

```python
median_age_men = df_1[df_1.Sex == 1]["Age"].median()
median_age_women = df_1[df_1.Sex == 0]["Age"].median()
```

```python
df_1.loc[(df_1.Age.isnull()) & (df_1.Sex == 0), "Age"] = median_age_women
df_1.loc[(df_1.Age.isnull()) & (df_1.Sex == 1), "Age"] = median_age_men
```

```python
df_1.isnull().sum()
```

```python
df_1.dropna(inplace =True)
```

```python
df_1.isnull().sum()
```

```python
df_1
```

```python
df_1.Age = (df_1.Age - min(df_1.Age) )/(max(df_1.Age)- min(df_1.Age))
df_1.Fare = (df_1.Fare - min(df_1.Fare) )/(max(df_1.Fare)- min(df_1.Fare))
```

```python
df_1.describe()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_1.drop(['Survived' ], axis = 1), df_1.Survived, test_size = 0.2, random_state = 0, stratify = df_1.Survived)
```

```python
clf = LogisticRegression()
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
sns.heatmap(cm, annot= True)
```
